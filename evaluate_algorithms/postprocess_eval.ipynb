```python
import os


for file in os.listdir('eval_category'):
    cat = file.rsplit('_',1)[0]
    path = os.path.join('eval_category', file)
    print(cat)
    df = pd.read_csv(path, index_col=0)
    if cat!='file_num' and not file.startswith('.ipynb'):
        df['num'] = df['num'].div(d[cat]).round(2)
    
        
        df_sorted = df.sort_values(by='f1 score', ascending=False)
        df_sorted.to_csv(path)
        print(df.sort_values(by='f1 score', ascending=False))
```

```python
for file in os.listdir('eval_category'):
    if file!='file_num_eval.csv':
        path = os.path.join('eval_category', file)
        df = pd.read_csv(path, index_col=0)
        df = df[['annotation error', 'precision', 'recall', 'f1 score', 'hausdorff',
           'rand index', 'gold num', 'predicted num', 'num']]
        print(df.columns)
        df.to_csv(path)
```
