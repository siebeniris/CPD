```python
def get_uid_sentence(file):
    uids = list()
    uid_sentence = dict()
    with open(file) as reader:
        data = json.load(reader)
    count =0
    for idx, l in data["sentences"].items():
        # idx : rupture point.
        for elem in l :
            # date, uid, lemma, sentence, sentiment, cpt.
            d = ast.literal_eval(elem[2])
            uid_sentence[count] = (elem[0], elem[1], d, elem[3], elem[4], idx)
            uids.append(elem[1])
            count+=1
    dates = data["dates_periods"]
    
    return uid_sentence, uids, dates
```

```python
def get_new_json_content(filename, category, file_id):
    review_list =[]
    if category=='renovation-related_room':
            filepath = os.path.join(data_dir, "renovation-related", "room", "json_file", filename)
            cats = ['renovation', 'room']
    else:
        filepath = os.path.join(data_dir, category, "json_file", filename)
        cats = [category]
    
    uid_sentence, uids, dates = get_uid_sentence(filepath)
    hotel_id = filename.split('#')[0]
    keywords = list(flatten([topics[x] for x in cats]))
    # load the csv_file
    csv_filename= filename.replace('.json', '')
    csv_filepath = os.path.join(sc_dir, csv_filename)
    print(csv_filepath)
    
    name= '_'.join(cats+[hotel_id])
    
    
    df = pd.read_csv(csv_filepath)
    
    rows = df[df['uid'].isin(uids)]

    # regroup the sentences into reviews.
    groups = rows.groupby('uid')['sentence'].apply(list).reset_index(name ='new')

    uid_sents_list =dict()
    for uid, new in zip(groups.uid, groups.new ):
        uid_sents_list[uid] = new

    for idx, sentence in uid_sentence.items():
        # uid, sent, date, cpt, category = sentence
        date, uid, lemma, sent, sentiment, cpt = sentence
        sent_list = uid_sents_list[uid]
        try:
            review = '. '.join(sent_list)+'.'

            sent_lower= sent.lower()
            hashtags = list(set(lemma).intersection(set(keywords)))

            offset = (review.find(sent), review.find(sent)+len(sent)+1)

            # print(sent[offsets[0][0]:offsets[0][1]])

            review_list.append({
                "id":idx,
                "uid":uid,
                "review":review, 
                "offset_sent":offset,
                "hashtags":hashtags,
                "date":date,
                "cpt": cpt, 
            })

        except Exception:
            print(Exception)

    return {"id": file_id, "name":name, "categories":cats, "dates_periods":dates, "review_list": review_list}


```
