```python
# should there be a/b/c in one time series. then record it.
ls_dfs = []
cpts_nr = 0
cpts_dict =defaultdict(dict)
for key, group in sents_df.groupby('fileId'):
    cpcs = list(set(group['cpt_id'].to_list()))
    cpc_nr = len(cpcs)
#     print(cpcs)  #['1_1', '1_0']
    if key not in cpts_dict:
        cpts_dict[key] = {}
    
    for cpt_id, cpt_label in zip(group['cpt'], group['cpt_label']):
        if cpt_id not in cpts_dict[key]:
            cpts_dict[key][cpt_id] = cpt_label
            
    cpts_nr += cpc_nr-1
    
#     print(cpc_nr) #2
    if cpc_nr>1:
        ls_dfs.append(group)
```

```python
import numpy as np

def max_dist_idx(number, ls):
    distances = [abs(number-x) for x in ls]
    return np.argmax(distances)
```

```python
gold_df['argmin_max']=""
```

```python
for key, group in gold_df.groupby('cpt_id'):
    label = group['cpt_label'][0]
    mean_sentiment = group['sentiment_mean'][0]
    if label=='a':
        
        idx_min = group['sentiment'].idxmin(1)
        gold_df.at[idx_min, 'argmin_max'] = True
    if label=='b':
        idx_max = group['sentiment'].idxmax(1)
        gold_df.at[idx_max, 'argmin_max']= True
    
    if label=='c' or label=='d':
        max_id = max_dist_idx(mean_sentiment, group['sentiment'].to_list())
        row = group.iloc[max_id]
        fileid = int(row.fileId)
        iid = int(row.iId)
        idx = str(fileid)+'_'+str(iid)
        gold_df.at[idx, 'argmin_max']=True
```

```python
cpts_d = defaultdict(dict)
for key, group in gold_df.groupby('fileId'):    
    cpt_list= group['cpts'].to_list()[0]
    cpts = sorted(list(set(group['cpt'].to_list())))
    cpt_label =[]
    for k, g in group.groupby('cpt_id'):
        cpt_label.append(g['cpt_label'][0])
    label_dict = dict(zip(cpts, cpt_label))
    category= group['category'].to_list()[0]
    hotel_id = group['hotel_id'].to_list()[0]
    
    cpts_d[key]={
        "cpts": cpt_list,
        "cpt_dict":label_dict,
        'category':category, 
        'hotel_id':hotel_id
    }
    
```

```python
cpts_d[0]
```

```python
for fileId, cpt_dict in cpts_d.items():
    cpts = json.loads(cpt_dict['cpts'])
    cpt_list= cpt_dict['cpt_dict']
    print(cpt_dict)
    periods_cands = list(zip(cpts[:-1], cpts[1:]))
    cpts_ = [x for x,y in cpt_list.items() if y!='c']
    labels=[y for x,y in cpt_list.items() if y!='c']
    valid_periods = [periods_cands[x] for x in cpts_]
    periods= [0] + [y for x,y in valid_periods]
    if periods[-1]!= cpts[-1]:
        periods += [cpts[-1]]
    print(valid_periods)
    print(periods)
    if labels[-1]=='d':
        labels = labels[:-1]
    if len(labels)>0:
        cpts_d[fileId]['valid_cpts']= periods
        cpts_d[fileId]['valid_labels'] = labels
    print(labels)
    print('*'*20)
```

```python
new_cpt_d = {}
# should exclude those only with c and d
for idx, cpts in cpts_d.items():
    cs = set(cpts['cpt_dict'].values())
    if cs != {'c','d'}:
        new_cpt_d[idx]= cpts
```
