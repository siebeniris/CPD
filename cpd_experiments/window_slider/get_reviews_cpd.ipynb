```python

bkps_filename = os.path.join(bkps_dir, filename+'.json')
with open(bkps_filename) as file:
    bkps_dict = json.load(file)

dates = bkps_dict['dates']
scores = bkps_dict['scores']
bkps = bkps_dict['bkps']
periods = list(zip(bkps[:-1], bkps[1:]))



print(bkps)
period_ids = {}
period_ids[0] = (0, bkps[0])
count=1
for z in periods:
    x,y = z
    period_ids[count]=(x,y)
    count+=1
print(period_ids)


period_scores ={}
period_scores[(0, bkps[0])] = np.mean(scores[:bkps[0]])

reno_df = df[df['renovation']==True]
reno_df.loc[(reno_df['date']<=dates[period_ids[0][1]]), 'point'] = 0

for idx, period in period_ids.items():
    if idx >=1:
        start, end = period
        period_scores[(start, end)]= np.mean(scores[start:end])
        reno_df.loc[(reno_df['date']> dates[start]) & (reno_df['date']<= dates[end-1]), 'point' ]=idx

        
        

sentences = {}




uids = {}
sentences ={}
print(df.head(2))
df = df[df['renovation']==True]
print('==='*40)
print('going up')

df0 = df[df['point']==0][df['polarity']<1]
uids[0] =list(set( df0.uid.to_list()))
sentences[0] = list(zip(df0.sentence, df0.date))

print('==='*40)
print('high')
df1 = df[df['point']==1][df['polarity']==1]
uids[1]= list(set( df1.uid.to_list()))
sentences[1] = list(zip(df1.sentence, df1.date))

print('==='*40)
print('going down')
df2 = df[df['point']==2][df['polarity']>-1]
uids[2]= list(set( df2.uid.to_list()))
sentences[2] = list(zip(df2.sentence, df2.date))

print('==='*40)
print('low')
df3 = df[df['point']==3][df['polarity']==-1]
uids[3]= list(set( df3.uid.to_list()))
sentences[3] = list(zip(df3.sentence, df3.date))



period_scores
```
